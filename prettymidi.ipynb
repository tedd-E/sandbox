```python
import pretty_midi

# Define the path to your MIDI file
midi_file_path = 'bohemian_rhapsody.mid'

# Load the MIDI file
midi_data = pretty_midi.PrettyMIDI(midi_file_path)

# Access different components of the MIDI data
print(f"Number of Instruments: {len(midi_data.instruments)}")
print(f"Instrument Data: {midi_data.instruments}")

# instrument = midi_data.instruments[0]
for instrument in midi_data.instruments:
    for note in instrument.notes:
        print(instrument)
        print(note)
```
